# **Q3.1.** **For temperature predictions**, the model performs well across all metrics. **R² = 0.688 **indicates that a substantial proportion of variance is explained. **MAE = 0.8 and RMSE = 1.0** show that prediction errors are small in absolute terms. **MAPE = 5.8%** confirms that the relative prediction error is low. **For rainfall predictions**, performance is significantly weaker. **R² = 0.231** indicates poor explanatory power. **MAE = 1.2 and RMSE = 2.608** suggest larger prediction errors, and the higher RMSE compared to MAE indicates the presence of larger individual errors. **MAPE becomes extremely large **because rainfall contains zero values, making percentage-based error unstable due to division by zero.

Overall, I learn that:

	•	R² measures how much variance is explained by the model.

	•	MAE (Mean Absolute Error) measures the average absolute difference between predicted and actual values.
It is expressed in the same units as the target variable, making it directly interpretable (e.g., degrees, millimeters). It reflects the typical size of the prediction error.

	•	RMSE (Root Mean Squared Error) measures the square root of the average squared differences between predicted and actual values.
Because the errors are squared before averaging, larger errors are penalized more heavily. RMSE is therefore more sensitive to outliers and large prediction mistakes than MAE.

	•	MAPE (Mean Absolute Percentage Error) measures the average absolute error as a percentage of the true values.
It expresses prediction error in relative terms, making it easy to interpret as a percentage. However, it becomes unstable or undefined when true values are zero, since it involves division by the actual value.

```python
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error

temp   = np.array([12,13,15,12,11,11,17,13,12,14])
t_pred = np.array([12,12,14,13,12,11,15,12,12,13])

rain   = np.array([0,0,5,7,1,1,0,8,0,4])
r_pred = np.array([0,1,4,7,1,0,0,0,1,4])
```

```python
print("TEMP")
print("R²:", round(r2_score(temp, t_pred), 3))
print("MAE:", round(mean_absolute_error(temp, t_pred), 3))
print("MAPE:", round(mean_absolute_percentage_error(temp, t_pred), 3))
print("RMSE:", round(np.sqrt(mean_squared_error(temp, t_pred)), 3))
```

```python
print("\nRAIN")
print("R²:", round(r2_score(rain, r_pred), 3))
print("MAE:", round(mean_absolute_error(rain, r_pred), 3))
print("MAPE:", round(mean_absolute_percentage_error(rain, r_pred), 3))
print("RMSE:", round(np.sqrt(mean_squared_error(rain, r_pred)), 3))
```

# **Q4.** **(3). They sometimes tell us the relative importance of the features in the model.**

Explanation:

Regression coefficients (β) indicate the change in the target variable for a one-unit change in a feature, holding other variables constant.

However, they do not always reflect relative importance because:
	•	Coefficients depend on the scale of the variables.
	•	Features measured in different units are not directly comparable.
	•	Multicollinearity can distort coefficient magnitudes.

Only when features are properly normalized and not highly correlated can coefficients meaningfully reflect relative importance.

Therefore, the correct statement is that they sometimes tell us the relative importance — not always.

# **Q5.1.** Simple Linear Regression Model:
	•	Slope (β₁) ≈ 25.14
	•	Intercept (β₀) ≈ 131.28

Interpretation

The slope (25.14) means that for each additional hour of sunshine, ice cream sales increase by approximately 25 units on average.

The intercept (131.28) represents the predicted sales when there are 0 hours of sunshine.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([4.3,6.9,0.0,10.4,5.2,1.8,8.0,9.2,2.1,1.2]).reshape(-1, 1)
y = np.array([224,271,130,419,230,184,362,351,196,180])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print("Mean x:", x_mean)
print("Mean y:", y_mean)
```

```python
numerator = np.sum((x.flatten() - x_mean) * (y - y_mean))
denominator = np.sum((x.flatten() - x_mean)**2)

beta_1 = numerator / denominator

print("Slope (beta_1):", beta_1)
```

```python
beta_0 = y_mean - beta_1 * x_mean
print("Intercept (beta_0):", beta_0)
```

```python
model = LinearRegression()
model.fit(x, y)

print("Sklearn slope:", model.coef_[0])
print("Sklearn intercept:", model.intercept_)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# Data
x = np.array([4.3,6.9,0.0,10.4,5.2,1.8,8.0,9.2,2.1,1.2]).reshape(-1,1)
y = np.array([224,271,130,419,230,184,362,351,196,180])

# Fit model
model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

# Metrics
r2 = r2_score(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))
mape = mean_absolute_percentage_error(y, y_pred)

print("R²:", round(r2,3))
print("MAE:", round(mae,2))
print("RMSE:", round(rmse,2))
print("MAPE:", round(mape,3))
```
